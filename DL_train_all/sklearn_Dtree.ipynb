```python
#sklearn 决策树+ 可视化 
from sklearn.datasets import load_iris
from sklearn import tree
import graphviz
import numpy as np 
from sklearn.model_selection import train_test_split
```

```python
def get_data(filename):
    f = open(filename,'r',encoding='utf-8')
    all_data = f.readlines()
    #字符
    X = [[[0]for i in range(3)]for j in range(len(all_data))]
    y = [[0]for i in range(len(all_data))]
    for i in range(len(all_data)):
        line = all_data[i].strip().split(',')  # 以逗号为分割符拆分列表
        X[i][0] = int(line[0])
        X[i][1] = int(line[1])
        X[i][2] = int(line[2])
        y[i] = int(line[-1])-1
    
    return X,y

Xdata,ydata = get_data('./data/haberman.data')
```

```python
X_train,X_test,y_train,y_test = train_test_split(Xdata,ydata,test_size = 0.3)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
# print(clf.score(X_test,y_test))
# print(clf.predict(X_test))
```
