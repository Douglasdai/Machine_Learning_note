```python
import pandas as pd 
import csv
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
```

```python
data = pd.read_csv('D:\ML_data_sql\info.csv',header=None,names=['id','name','year','rating','genre','director','actor','other actor','area'])
data = data[['id','genre']]
data.head()
data.info()
```

```python
import re
genre = set()
deals = data['genre']
for val in deals:
    str = val
    str2 = str.split('/')
    for val2 in str2:
        genre.add(val2)
genre.remove('')
#generate categroy
categroy = [[]for i in range(len(genre))]
i=0
for val in genre:
    categroy[i] = val
    i+=1
```

```python
categroy = [[]for i in range(len(genre))]
i=0
for val in genre:
    categroy[i] = val
    i+=1
```

```python
score =[[] for i in range(len(genre))]
print(len(score))
i=0
for val in categroy:
    for deal in deals:
        cls = deal.split('/')
        cls.remove('')
        if val in cls:
            score[i].append(1)
        else:
            score[i].append(0)
    i+=1
#一列
print(len(score[0]))    
# s = np.sum(score[-1])
# if s==0:
#     print('adf')
from pandas.core.frame import DataFrame
```

```python
save_data = data['id']
print(len(save_data))
pscore = DataFrame(score)
# pscore.T.head()
save_data = pd.concat([save_data,pscore.T],axis=1,ignore_index=True)
if save_data.shape[1]!= len(categroy):
    categroy.insert(0,'id')
# print(categroy)
save_data.columns = categroy
# save_data[2:] = save_data[2:].astype(int)
save_data.head(-1)
# save_data.info()
```
